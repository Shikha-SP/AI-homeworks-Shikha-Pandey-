```python
#problem 4
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

A_inv = np.linalg.inv(A)       
I = np.dot(A, A_inv)             
print("A * A^-1 =\n", I)

AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB =\n", AB)
print("BA =\n", BA)
print("AB == BA ?", np.array_equal(AB, BA))

AB_T = np.transpose(AB)
BT_AT = np.dot(np.transpose(B), np.transpose(A))
print("\n(AB)^T =\n", AB_T)
print("B^T * A^T =\n", BT_AT)
print("(AB)^T == B^T * A^T ?", np.array_equal(AB_T, BT_AT))


coeff_matrix = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3, 1, -1]])

B_vector = np.array([-1, -3, 9])

X = np.dot(np.linalg.inv(coeff_matrix), B_vector)
print("\nSolution for [x, y, z] =", X)
```
